```python
import stats
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/kjams/Desktop/pyVis/data/nyc.csv')
```

```python
jedi = stats.formulas(df)
jedi.set_x_y('Poverty','Income')
jedi.pdf_linearBinning('Poverty')
jedi.pdf_log_binning('Poverty')
jedi.ctl('Poverty',1000)
jedi.corr('Poverty','Income')
print(df['Poverty'].corr(df['Income']))
```

```python
jedi.init_knn(5, ['Poverty', 'Income'])
```

```python
print(jedi.graph[0])
node = jedi.graph[0]
print('x:',node.x,' y:',node.y, ' idx:', node.idx)
print('names:', node.names)
print('cols:', jedi.cols)
print(node.entropy('Unemployment'))
```
